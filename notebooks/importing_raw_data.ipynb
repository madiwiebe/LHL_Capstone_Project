Reading and filtering metadata file:

```python
# read station metadata
station_metadata = pd.read_csv(r"D:\School\LHL\capstone_project_data\raw_datasets\station-metadata-by-history.csv")
station_metadata
```

```python
# specify only the relevant networks and variables
station_metadata_filtered = station_metadata.loc[station_metadata["Network Name"].isin(['ENV-AQN','FLNRO-FERN','FLNRO-WMB','MoTIm','MoTIe','EC_raw','BCH','CRD','UNBC_CAM','MVan']), ["Network Name","Native ID","Latitude","Longitude","Elevation (m)"]]
# rename columns
station_metadata_clean = station_metadata_filtered.rename(columns={"Network Name": "network_id", "Native ID": "station_id", "Latitude": "latitude", "Longitude": "longitude", "Elevation (m)": "elevation"})
# replace network_id values containing "-" with "_"
station_metadata_clean.loc[:,("network_id")] = station_metadata_clean.loc[:,("network_id")].replace("-","_", regex=True)
# reset index
station_metadata_clean = station_metadata_clean.reset_index(drop=True)
station_metadata_clean
```

```python
# save clean dataframe to csv
station_metadata_clean.to_csv('../data/metadata/station_metadata_clean.csv', index=False)
```
